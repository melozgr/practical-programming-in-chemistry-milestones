```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def plot_regression(x, y):
    # Reshape x to meet the requirements of the sklearn library (2D array for features)
    x = np.array(x).reshape(-1, 1)
    y = np.array(y)

    # Create a linear regression object and fit it to the data
    model = LinearRegression()
    model.fit(x, y)

    # Predict y values across the given x range to plot the regression line
    y_pred = model.predict(x)

    # Calculate the R² value
    r2 = r2_score(y, y_pred)

    # Plotting the original data points
    plt.scatter(x, y, color='blue', label='Data Points')

    # Plotting the regression line
    plt.plot(x, y_pred, color='red', linewidth=2, label='Regression Line')

    # Adding the R² value in the plot
    plt.text(x.min(), y.max(), f'R² = {r2:.2f}', fontsize=12, verticalalignment='top')

    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.title('Linear Regression Analysis')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
x_coords = [1, 2, 3, 4, 5]
y_coords = [1, 4, 5, 4, 5]
plot_regression(x_coords, y_coords)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def plot_regression(x, y):
    # Reshape x to meet the requirements of the sklearn library (2D array for features)
    x = np.array(x).reshape(-1, 1)
    y = np.array(y)

    # Create a linear regression object and fit it to the data
    model = LinearRegression()
    model.fit(x, y)

    # Predict y values across the given x range to plot the regression line
    y_pred = model.predict(x)

    # Calculate the R² value
    r2 = r2_score(y, y_pred)

    # Plotting the original data points
    plt.scatter(x, y, color='blue', label='Data Points')

    # Plotting the regression line
    plt.plot(x, y_pred, color='red', linewidth=2, label='Regression Line')

    # Adding the R² value in the plot at a more visible location
    plt.text(x.min(), y.max(), f'R² = {r2:.2f}', fontsize=12, verticalalignment='bottom', horizontalalignment='left')

    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.title('Linear Regression Analysis')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
x_coords = [1, 2, 3, 4, 5]
y_coords = [2, 4, 5, 4, 5]
plot_regression(x_coords, y_coords)
```
